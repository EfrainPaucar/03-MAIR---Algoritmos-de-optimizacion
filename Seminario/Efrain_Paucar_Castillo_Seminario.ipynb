1) Elección de poblaciones homogéneos

Descripción del Problema:
Una productora ganadera nos encarga la tarea de seleccionar grupos de
terneros para aplicar 3 tratamientos diferentes. Para cada uno de los
tratamientos debemos seleccionar 3 grupos de terneros separados por sexos
que sean lo mas homogéneos posible en peso para que en los resultados del
tratamiento influya lo menos posible el peso del animal. Disponemos de una
población de N animales entre machos y hembras

Se solicita diseñar un algoritmo para conseguir una agrupación que cumpla de la
mejor manera posible las especificaciones de la productora.
******************************************************************************

### Respuestas:
¿Cuantas posibilidades hay sin tener en cuenta las restricciones?

    Respuesta: Combinación de N a 3.
    
¿Cual es la estructura de datos que mejor se adapta al problema? 

    Respuesta: La data mínima requerida es una lista de con columnas de: sexo y peso. Sería mejor
               cuando está ordenado por sexo y peso.
    
¿Cual es la función objetivo?

    Respuesta: Encontrar que se encuentran en cada rango (Inferior,Superior), para que sean homogeneos en 3 rangos.
    
¿Es un problema de maximización o minimización?

    Respuesta: Minimizar la distancia entre los integrantes en cada grupo.


```python
#Generando datos aleatorios con media = 35 y desviación estandar = 3.1
import scipy.stats as ss
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

#columna 0: sexo
#columna 1: peso
#columna 2: grupo
nodo = dict.fromkeys(['sexo','peso', 'grupo'])
df = pd.DataFrame(columns=nodo)

x = np.random.normal(35, 3.1, size=100)
#x = (x - x.mean()) / x.std()

#sexo: 0 hembra y 1 macho
sexo = np.random.randint(2, size=100)

for i in range(x.size):
    df.at[i,'sexo'] = sexo[i]
    df.at[i,'peso'] = x[i]
    df.at[i,'grupo'] = 0

#número de grupos homogéneos
N = 3 
```

```python
def gruposhomogeneos(df,N):
    #División en grupos por sexo
    
    #Mínimo-Maximo Hembra
    PesoMinH = min(df.peso[df.sexo==0])
    PesoMaxH = max(df.peso[df.sexo==0])
    #Mínimo-Maximo Macho
    PesoMinM = min(df.peso[df.sexo==1])
    PesoMaxM = max(df.peso[df.sexo==1])

    #inicilización límite inferior de grupos
    LimiteH = []
    LimiteM = []
    LimiteH.append(PesoMinH)
    LimiteM.append(PesoMinM)

    for i in range(N):
        LimiteH.append(PesoMinH+(i+1)*(PesoMaxH-PesoMinH)/N)
        LimiteM.append(PesoMinM+(i+1)*(PesoMaxM-PesoMinM)/N)
    #Límite superior
    LimiteH[N]=PesoMaxH+0.001
    LimiteM[N]=PesoMaxM+0.001
    
    for i in range(len(df)):
        for j in range(N):
            #hembras
            if (df.at[i,'sexo']==0):
                if (df.at[i,'peso']>=LimiteH[j]) and (df.at[i,'peso']<LimiteH[j+1]):
                    df.at[i,'grupo']=j+1
            else: #machos
                if (df.at[i,'peso']>=LimiteM[j]) and (df.at[i,'peso']<LimiteM[j+1]):
                    df.at[i,'grupo']=N+j+1
    return df.sort_values(['peso'])
```

```python
df_grupos=gruposhomogeneos(df,N)
```
